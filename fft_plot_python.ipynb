```python
# Computing the FFT and plotting Frequency Spectrum in Python w/ numpy and matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = [15, 6]
```

```python
# 1. construct a time signal
Fs = 2000 # sample freq of 2kHz
tstep = 1 / Fs  # sample time interval

f0 = 100  # signal frequency
f1 = 300

N = int(10*Fs / f0)   # number of samples

t = np.linspace(0, (N-1)*tstep, N) # time steps
fstep = Fs / N # freq interval
f = np.linspace(0, (N-1)*fstep, N) # freq steps

y = 3 * np.sin(2 * np.pi * f0 * t) + 5 * np.sin(2 * np.pi * f1 * t)





# 2. perform fft
X = np.fft.fft(y)
X_mag = np.abs(X) / N  # only interested in magnitude of fft output, so use abs value, then normalize with # samples

f_plot = f[0:int(N/2+1)]   # nyquist thm says we get plot frequencies only less than half of our sampling rate
X_mag_plot = 2 * X_mag[0:int(N/2+1)]
X_mag_plot[0] = X_mag_plot[0] / 2   # Note: DC component noes not need to be multiplied by 2 for unexplained reason


# 3. plot
fig, [ax1, ax2] = plt.subplots(nrows=2, ncols=1)
ax1.plot(t, y, '.-')
ax2.plot(f_plot, X_mag_plot, '.-')

ax1.set_xlabel("time (s)")
ax2.set_xlabel("freq (Hz)")

ax1.set_xlim(0, t[-1])
ax2.set_xlim(0, f_plot[-1])

ax1.grid()
ax2.grid()
plt.tight_layout()
plt.show()
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import scipy.fftpack

# Number of sample points
N = 10000

# Sample spacing
T = 1.0 / 51.2

x = np.linspace(0.0, N*T, N)
y = np.sin(2.0 * 2.0 * np.pi * x) + 0.5*np.sin(3.0 * 2.0 * np.pi * x)
yf = scipy.fftpack.fft(y)
xf = np.linspace(0.0, 1.0/(2.0*T), int(N/2))

fig, ax = plt.subplots()
ax.plot(xf, 2.0/N * np.abs(yf[:int(N/2)]))
plt.show()
```
